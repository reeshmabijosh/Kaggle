```python
#1587 nan values replaced with oter
def Hospital_admit(row):  

    if row['hospital_admit_source']=='Emergency Department':
        return 'Emergency Department'
    elif row['hospital_admit_source']=='Operating Room':
        return 'Operating Room'
    elif row['hospital_admit_source']=='Floor':
        return 'Floor'
    elif row['hospital_admit_source']=='Direct Admit':
        return 'Direct Admit'
    elif row['hospital_admit_source']=='Recovery Room':
        return 'Recovery Room'
    elif row['hospital_admit_source']=='Other Hospital':
        return 'Other Hospital'
    elif row['hospital_admit_source']=='Acute Care/Floor':
        return 'Acute Care/Floor'
    elif row['hospital_admit_source']=='Step-Down Unit (SDU)':
        return 'Step-Down Unit (SDU)'
    elif row['hospital_admit_source']=='PACU':
        return 'PACU'    
    else:
        return 'other'
    
df['hospital_admit_source'] = df.apply(lambda row: Hospital_admit(row), axis=1)
```

It is one of the most important evaluation metrics for checking any classification modelâ€™s performance.ROC is a probability curve and AUC represents the degree or measure of separability. It tells how much the model is capable of distinguishing between classes. Higher the AUC, the better the model is at predicting 0s as 0s and 1s as 1s.
